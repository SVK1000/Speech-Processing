```python
!pip install soundfile simpleaudio
!pip install librosa scipy

import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import librosa
import simpleaudio as sa

file_path = "/content/LJ018-0075.wav"
data, sample_rate = sf.read(file_path)

print("Sample Rate:", sample_rate)
print("Number of Samples:", len(data))
print("Duration:", len(data) / sample_rate, "seconds")

# Plotting the waveform of the speech signal
plt.figure(figsize=(10, 4))
plt.plot(data)
plt.title('Waveform of Speech Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()

# Slicing the first 2 seconds
sliced_data = data[:2 * sample_rate]

plt.figure(figsize=(10, 4))
plt.plot(sliced_data)
plt.title('Sliced Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()

# Normalization
normalized_data = sliced_data / np.max(np.abs(sliced_data))

plt.figure(figsize=(10, 4))
plt.plot(normalized_data)
plt.title('Normalized Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()

# Amplification
amplified_data = data * 2

plt.figure(figsize=(10, 4))
plt.plot(amplified_data)
plt.title('Amplified Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()

# De-amplification
deamplified_data = data * 0.5

plt.figure(figsize=(10, 4))
plt.plot(deamplified_data)
plt.title('De-Amplified Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()

# Up-sampling
up_sampled_data = librosa.resample(y=data, orig_sr=sample_rate, target_sr=sample_rate * 2)

plt.figure(figsize=(10, 4))
plt.plot(up_sampled_data)
plt.title('Up-Sampled Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()

# Down-sampling
down_sampled_data = librosa.resample(y=data, orig_sr=sample_rate, target_sr=sample_rate // 2)

plt.figure(figsize=(10, 4))
plt.plot(down_sampled_data)
plt.title('Down-Sampled Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.show()
```
